```python
dwarf = "segue_1"

for i, run in enumerate(tqdm(printRunList(dwarf, log_only=True))):
    pars = ResponseFunction.getParsFromLog(dwarf, int(run))
    
    irf = ResponseFunction.EventDisplay(dwarf, int(run), mode="0", verbose=False)
    irf.zn = 90-pars["el"]
    irf.az = pars["az"] if pars["az"]> 0 else pars["az"]+360
    irf.noise = pars["ped"]
    irf.offset = 0.5

    irf.SetInterpolator()
    exp, irf.Exp = irf.GetExposure(dwarf, int(run))
    irf.EA = irf.GetEffectiveArea(verbose=False)
    irf.Edisp = irf.GetEnergyDispersionMatrix(verbose=False)
    irf.PSF = irf.GetPointSpreadFunction(verbose=False)
    irf.__exportIRF__()
```

```python
hSig1D = TH1D("SignificanceDistribution", "Significance Distribution", 200, -5, 5)
for i in range(400):
    for j in range(400):
        x = hSig.GetXaxis().GetBinCenter(i)
        y = hSig.GetXaxis().GetBinCenter(j)
        r = np.sqrt(x**2+y**2)
        sig = hSig.GetBinContent(i,j)
        if r < 1.5 and sig !=0.0:
            hSig1D.Fill(sig)

c=TCanvas()
hSig1D.Draw()
c.Draw()
```

```python


zn_az = {}
for dwarf in ["segue_1", "draco", "ursa_minor", "bootes"]:
    zn_az[dwarf] = []
    for i, run in enumerate(tqdm(printRunList(dwarf, log_only=True))):
        pars = ResponseFunction.getParsFromLog(dwarf, int(run))

        zn_az[dwarf].append([90-pars["el"], pars["az"] if pars["az"]> 0 else pars["az"]+360, pars["ped"]])
    zn_az[dwarf] = np.asarray(zn_az[dwarf])    
```
