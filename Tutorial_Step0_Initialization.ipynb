```python
dwarf = "segue_1"

for i, run in enumerate(tqdm(printRunList(dwarf, log_only=True))):
    pars = ResponseFunction.getParsFromLog(dwarf, int(run))
    
    irf = ResponseFunction.EventDisplay(dwarf, int(run), mode="0", verbose=False)
    irf.zn = 90-pars["el"]
    irf.az = pars["az"] if pars["az"]> 0 else pars["az"]+360
    irf.noise = pars["ped"]
    irf.offset = 0.5

    irf.SetInterpolator()
    exp, irf.Exp = irf.GetExposure(dwarf, int(run))
    irf.EA = irf.GetEffectiveArea(verbose=False)
    irf.Edisp = irf.GetEnergyDispersionMatrix(verbose=False)
    irf.PSF = irf.GetPointSpreadFunction(verbose=False)
    irf.__exportIRF__()
```
